```python
#Reading the actual data file
text = ""
with open('data/englishfairytales/englishfairytales.txt', 'r') as file:
    lines = file.readlines()
    for line in lines:
        text += line
```

```python
#Split up into stories by splitting on new line characters
chunks = text.split('\n\n\n\n\n\n\n\n')
```

```python

#Get rid of chunks -4 -> -1 and chunks 0 -> 3
chunks = chunks[4:-4]
```

```python

import re
from transformers import pipeline

#Function that returns a dictionary of sentences and sentiments for each sentence in the story
def classify_chunk(chunk:str) -> list:
    output = []
    classifier = pipeline(task="text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)
    lines = chunk.strip().split("\n")  # Split by newline characters to get lines

    sentences = [sentence for line in lines for sentence in re.split(r'(?<=[.])(?<!Mr\.|rs\.|MR\.|RS\.)\s+', line)] #Don't split on Mr., Mrs., MR., or MRS.
    print(sentences)
    sentiments = classifier(sentences)
    for index, sentence in enumerate(sentences):
        if len(sentence) > 1: #No '' or '"' sentenvce
            output.append({"Sentence": sentence, "Sentiment":sentiments[index]})
    return output
```

```python
stories = []
for chunk in chunks: 
    output = classify_chunk(chunk)
    stories.append({"title":output[0]["Sentence"], "output":output})
print(stories)
```

```python
import json

file_path = "data/englishfairytales/semantic_data.json"

with open(file_path, "w") as json_file:
    json.dump(stories, json_file)
```
